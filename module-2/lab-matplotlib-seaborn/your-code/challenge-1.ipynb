```python
x = np.arange(0,100)
#arange is equal to range
y = x*2
z = x**2
plt.plot(x,y,color='r',linestyle='--')
plt.plot(x,z,color='b',linewidth=0.4,marker='o')
```

#### Plot both `y=x^2` and `y=exp(x)` in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here

y=x*2
z=np.exp(x)

fig, ax = plt.subplots(1,2, figsize=(10,5))
ax[0].plot(x, y, x, z) #axis 0 draws in the first space
ax[0].set_title('squared expression')
ax[0].set(xlabel='x axis',ylabel='y axis')
ax[0].set_xscale= ('linear')


ax[1].plot(x, y, x, z)
ax[1].set_title('logaritmic expression')
ax[1].set(xlabel='x axis',ylabel='y axis')
ax[1].set_xscale= ('log')
```
